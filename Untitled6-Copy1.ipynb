```python
import matplotlib.pyplot as plt
import numpy as np

X1 = plt.imread("image1.png")
#plt.plot(X1)
#plt.show()

X2 = plt.imread("image2.png")
#plt.plot(X2)
#plt.show()
```

```python
K1 = np.vsplit(X1,len(X1))
K2 = np.vsplit(X2,len(X2))
```

```python
K1 = np.asarray(K1)
K2 = np.asarray(K2)
```

```python
x1 = np.hstack((K1))
x2 = np.hstack((K2))
X = np.vstack((x1, x2))
```

```python
X.shape
```

```python
def mean(X):
    E = np.zeros((len(X),1))
    for i in range (0, len(X)):
        total = 0
        for j in range (0, len(X[i])):
            total = total + X[i][j]
        mean = total / len(X[i])
        E[i][0] = mean
    return E

E = mean(X)
mean(X-E)

def sigma(X):
    a,b = X.shape
    temp = np.transpose(X)
    c,d = temp.shape
    sum = np.zeros((a,a))
    for i in range (0, c):
        x = np.zeros((d,1))
        for j in range(0, d):
            x[j][0] = temp[i][j]
        sum = sum + np.matmul(x, np.transpose(x))
    return sum / b

sigma(X-E)
d, e = np.linalg.eig(sigma(X-E))
print(d)
print(e)

def eigval_mat(d, I):
    for i in range (0, len(d)):
        I[i] = I[i] * d[i]
    return I

a, b = X.shape
D = eigval_mat(d, np.identity(a))
D

def pow_diag(D, n):
    for i in range (0, len(D)):
        for j in range(0, len(D[i])):
            if i == j:
                D[i][j] = D[i][j] ** n
    return D

pow_diag(D, -0.5)
D

V = np.matmul(np.matmul(e, D), np.transpose(e))
V

Z = np.matmul(V, X-E)
sigma(Z)
```

```python
def ortho(W):
    for i in range (0, len(W)):
        W[i] = W[i]/np.linalg.norm(W[i])
    return W

def kurtosis(Y):
    kur = np.zeros((len(Y),1)) #len(X)or len(Y)
    kur = mean(Y**4)-3*(mean(Y**2)**2)
    return kur

def update2(W): #W = 2x2 = [[w1T],[w2T]]
    for k in range(0, len(W)):
        #print('W before:',W)
        res = np.zeros((len(W),1))
        total = np.zeros((len(W),1))
        for i in range(0, len(X[0])):#z_w = 3142 X = 2 x 3142    
            D = np.zeros((1,1))
            E = np.zeros((1,1))
            K = np.zeros((2,1))
            z = np.zeros((len(W),1))
            for j in range(0, len(W)): #z_h = 2
                z[j][0] = Z[j][i] #Z=2x3142
            D = np.matmul(np.asarray([W[k]]),z)
            E = D ** 3
            #print(np.matmul(z,E))
            K = np.matmul(z,E)
            total = total + K
        res = total / float(len(X[0]))
        W[k] = np.transpose(res - np.transpose(float(3) * np.asarray([W[k]]))) #problem?
        if (W[k][0]<0):
            W[k] = -W[k]
        #print('W[k]:',W[k])
    #print('W after:',W,'\n')
    return W
        #print('res:',res)
            #C = np.transpose([Z_W[j]])
            #D = np.matmul(np.asarray([W[i]]),np.transpose([Z_W[j]]))
            #E = np.asarray(D ** 3)
        
        #print("res shape:",res.shape)
        #W[k] = np.transpose(res - np.transpose(3 * np.asarray([W[k]]))) #problem?
        #print('W shape:',W[k].shape)
        #print('np asarray shape:',np.transpose(np.asarray([3 * W[k]])), np.transpose(np.asarray([3 * W[k]])).shape)

def boushi(W):
    if all(np.abs(np.abs(W[0])-np.abs(W[1])) < np.array((1e-5,1e-5))):
        W[1] = np.random.rand(1,2)
```

```python
W = np.random.rand(2,2)
ortho(W)
Y1 = np.matmul(W, Z)
print(W)
update2(W)
boushi(W)
ortho(W)
Y2 = np.matmul(W, Z)
print('kurtY1_start:',kurtosis(Y1))
print('kurtY2:',kurtosis(Y2),'\n')
while any(np.abs(kurtosis(Y1)) < np.abs(kurtosis(Y2))):
    ortho(W)
    Y1 = np.matmul(W, Z)
    print('W:',W)
    update2(W)
    ortho(W)
    Y2 = np.matmul(W, Z)
    boushi(W)
    print('kurtY1:',kurtosis(Y1))
    print('kurtY2:',kurtosis(Y2),'\n')

#Y2 = Y
print('Y:',Y2)
print('kurtY_end:',kurtosis(Y2))
for i in range (0, len(Y2)):
    if(Y2[i][0]<0):
        Y2[i] = -Y2[i]
    plt.plot(Y2[i])
    plt.show()
```
