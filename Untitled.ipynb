```python
import matplotlib.pyplot as plt
```

```python
X1 = []
for line in open('dat1.txt', 'r'):
    values = [float(s) for s in line.split()]
    X1.append(values[0])

plt.plot(X1)
plt.show()
```

```python
X2 = []
for line in open('dat2.txt', 'r'):
    values = [float(s) for s in line.split()]
    X2.append(values[0])
    
plt.plot(X2)
plt.show()
```

```python
X3 = []
for line in open("dat1.txt", 'r'):
    values = [float(s) for s in line.split()]
    X3.append(values[0])

for line in open("dat2.txt", 'r'):
    values = [float(s) for s in line.split()]
    X3.append(values[0])
    
plt.plot(X3)
plt.show()
```

```python
def subtract(lst, num):
    for i in range (0, len(lst)):
        lst[i] = lst[i] - num
    return lst
```

```python
def mean_vectors(x1, x2):
    mean = []
    i = 0
    for i in range (0, len(x1)):
        mean.append((x1[i] + x2[i])/2)
    return mean
```
