```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X1 = []
for line in open('dat1.txt', 'r'):
    values = [float(s) for s in line.split()]
    X1.append(values[0])

plt.plot(X1)
plt.show()
```

```python
X2 = []
for line in open('dat2.txt', 'r'):
    values = [float(s) for s in line.split()]
    X2.append(values[0])
    
plt.plot(X2)
plt.show()
```

```python
def mean_vectors(x1, x2):
    mean = []
    i = 0
    for i in range (0, len(x1)):
        mean.append((x1[i] + x2[i])/2)
    return mean
```

```python
X3 = mean_vectors(X1, X2)
```

```python
def sub_vec(x1):
    res = []
    for i in range (0, len(x1)):
        res.append(x1[i] - X3[i])
    return res
```

```python
def eigval_mat(d, I):
    for i in range (0, len(d)):
        I[i] = I[i] * d[i]
    return I
```

```python
D_1 = eigval_mat(d1, D1)
D_1
```

```python
def pow_orto_mat(D, n):
    for x in D:
        for y in x:
            if D[x][y] != 0:
                D[x][y] = D[x][y] ** n
    return D
```
