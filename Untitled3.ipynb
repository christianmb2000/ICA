```python
D1 = np.asarray(np.identity(len(X1),complex))
D1
```

```python
def eigval_mat(d, I):
    for i in range (0, len(d)):
        I[i] = I[i] * d[i]
    return I
```

```python
D = eigval_mat(d, D1)
D
```

```python
def add_diag(D, n):
    for i in range (0, len(D)):
        for j in range(0, len(D[i])):
            if i == j:
                D[i][j] = D[i][j] + n
    return D
```

```python
small = 1e-323
```

```python
add_diag(D, small)
```

```python
def pow_diag(D, n):
    for i in range (0, len(D)):
        for j in range(0, len(D[i])):
            if i == j:
                D[i][j] = D[i][j] ** n
    return D
```

```python
pow_diag(D, -0.5)
```
