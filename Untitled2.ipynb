```python
def add_diag(D, n):
    for i in range (0, len(D)):
        for j in range(0, len(D[i])):
            if i == j:
                D[i][j] = D[i][j] + n
    return D
```

```python
def pow_diag(D, n):
    for i in range (0, len(D)):
        for j in range(0, len(D[i])):
            if i == j:
                D[i][j] = D[i][j] ** n
    return D
```
