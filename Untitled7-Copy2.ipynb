```python
import matplotlib.pyplot as plt
import numpy as np

import wave

# Read the wave file
mix_1_wave = wave.open('speechA1.wav','r')
mix_2_wave = wave.open('speechA2.wav','r')
from scipy.io import wavfile
sampling_rate, mix1 = wavfile.read('speechA1.wav')
sampling_rate, mix2 = wavfile.read('speechA2.wav')
plt.plot(mix1)
plt.show()
plt.plot(mix2)
plt.show()

X = np.vstack((mix1, mix2))
```

```python
def mean(X):
    E = np.zeros((len(X),1))
    for i in range (0, len(X)):
        total = 0
        for j in range (0, len(X[i])):
            total = total + X[i][j]
        mean = total / len(X[i])
        E[i][0] = mean
    return E

E = mean(X)
mean(X-E)

def sigma(X):
    a,b = X.shape
    temp = np.transpose(X)
    c,d = temp.shape
    sum = np.zeros((a,a))
    for i in range (0, c):
        x = np.zeros((d,1))
        for j in range(0, d):
            x[j][0] = temp[i][j]
        sum = sum + np.matmul(x, np.transpose(x))
    return sum / b

sigma(X-E)
d, e = np.linalg.eig(sigma(X-E))
print(d)
print(e)

def eigval_mat(d, I):
    for i in range (0, len(d)):
        I[i] = I[i] * d[i]
    return I

a, b = X.shape
D = eigval_mat(d, np.identity(a))
D

def pow_diag(D, n):
    for i in range (0, len(D)):
        for j in range(0, len(D[i])):
            if i == j:
                D[i][j] = D[i][j] ** n
    return D

pow_diag(D, -0.5)
D

V = np.matmul(np.matmul(e, D), np.transpose(e))
V

Z = np.matmul(V, X-E)
sigma(Z)
```
